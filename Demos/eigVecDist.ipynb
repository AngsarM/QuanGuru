```python
from qTools import *

from Cahit.Plotting.plottingSettings import plottingSet

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.logspace(-5, 0, num=50, base=10.0)
Dims = 200

COE = [EigenVectorDist(X, int(Dims), 1) for X in x]
CUE = [EigenVectorDist(X, int(Dims), 2) for X in x]
CSE = [EigenVectorDist(X, int(Dims), 4) for X in x]

ax = plt.gca()
ax.step(x, COE, 'r-', label=r'COE ($\beta = 1$)', linewidth=3)
ax.step(x, CUE, 'g-', label=r'CUE ($\beta = 2$)', linewidth=3)
ax.step(x, CSE, 'b-', label=r'CSE ($\beta = 4$)', linewidth=3)

ax = plottingSet(ax)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim([0.00001, 1])
ax.set_ylim([0.1, 10*Dims])

plt.show()
```
