# Single Qubit Evolution

Single qubit Hamiltonian is:

$\large\hat{H}=\frac{w_{q}}{2}\hat{\sigma}_{z}$

where 

$\large\hat{\sigma_{z}}:=|1 \times 1|-|0 \times 0|$ is the Pauli-z operator


$|\psi(0)\rangle=C_{0}(0)|0\rangle+C_{1}(0)|1\rangle$ s.t. $\left|C_{0}(0)\right|^{2}+\left|C_{1}(0)\right|^{2}=1$


Schrodinger equation $\frac{d}{d t}|\psi(t)\rangle=-i \hat{H}|\psi(t)\rangle$

$\frac{d}{d t} C_{0}(t)|0\rangle+\frac{d}{d t} C_{1}(t)|1\rangle=-\frac{i}{2} w_{q}(C_{0}(t) \underbrace{\hat{\sigma}_{z}|0\rangle}_{-|0\rangle}+C_{1}(t) \underbrace{\hat{\sigma}_{z}|1\rangle}_{|1\rangle})$

1) $\frac{d}{d t} C_{0}(t)=\frac{i}{2} w_{q} C_{0}(t) \quad \rightarrow \quad C_{0}(t)=C_{0}(0) e^{i \omega_{q} t/ 2} \quad\left|C_{0}(t)\right|^{2}=\left|C_{0}(0)\right|^{2}$

2) $\frac{d}{d t} C_{1}(t)=-\frac{i}{2} w_{1} C_{1}(t) \rightarrow C_{1}(t)=C_{1}(0) e^{-i \omega_{q} t / 2} \quad\left|C_{1}(t)\right|^{2}=\left|C_{1}(t)\right|^{2}$


Matrix representation $|1\rangle=\left[\begin{array}{l}1 \\ 0\end{array}\right] \&|0\rangle=\left[\begin{array}{l}0 \\ 1\end{array}\right] \Rightarrow \hat{\sigma_{z}}=\left[\begin{array}{cc}1 & 0 \\ 0 & -1\end{array}\right]$

$|\psi(0)\rangle=\left[\begin{array}{l}C_{0}(0) \\ C_{1}(0)\end{array}\right] ; \quad \frac{d}{d t}\left[\begin{array}{l}C_{1}(t) \\ C_{0}(t)\end{array}\right]=-i \frac{w}{2}\left[\begin{array}{ll}1 & 0 \\ 0 & -1\end{array}\right]\left[\begin{array}{l}C_{1}(t) \\ C_{0}(t)\end{array}\right]=\left[\begin{array}{l}-i \frac{i \omega_{q}}{2} C_{1}(t) \\ i \frac{\omega_{q}}{2} C_{0}(t)\end{array}\right]$

$\left[\hat{H}, \hat{\sigma}_{z}\right]=0 \Rightarrow\left\langle\hat{\sigma}_{z}\right\rangle$ is a constant of $\mathrm{motion}$

$|\psi(t)\rangle=C_{0}(0) e^{i w_{q} t / 2}|0\rangle+C_{1}(0) e^{-i \omega_{q} t / 2}|1\rangle$

$\langle\psi(t)|=c_{0}^{*}(0) e^{-i w_{q} t / 2}\langle 0|+C_{1}^{*}(0) e^{i \omega_{q} t / 2}\langle 1|$

$\left\langle\psi(t)\left|\hat{\sigma_{z}}\right| \psi(t)\right\rangle=\left|c_{0}(v)\right|^{2} \underbrace{\left\langle 0\left|\hat{\sigma}_{1}\right| 0\right\rangle}_{-1}+\left|C_{1}(0)\right|^{2} \underbrace{\left\langle 1\left|\hat{\sigma}_{1}\right| 1\right\rangle}_{1} +C_{0}^{*}(0) C_{1}(0) e^{-i \omega_{q} t} \overbrace{\left\langle 0\left|\hat{\sigma}_{z}\right| 1\right\rangle}^{0} + C_{0}(0) C_{1}^{*}(0) e^{i \omega_{q} t} \overbrace{\left\langle 1\left|\hat{\sigma}_{z}\right| 0\right\rangle}^{0}$

$\left\langle\hat{\sigma_{z}}(t)\right\rangle=\left|C_{1}(0)\right|^{2}-\left|C_{0}(0)\right|^{2} \quad \text { Spin polarization is conserved. }$

$\hat{\sigma}_{x}:=|1 \times 0|+|0 \times 1| \quad \& \quad \hat{\sigma}_{y}:=\frac{1}{i}(|1 \times 0|-|0 \times 1| \mid)=i(|0 \times 1|-|1 \times 0|)$

$\left\langle\psi(t)\left|\hat{\sigma_{i}}\right| \psi(t)\right\rangle=\left|c_{0}(0)\right|^{2}\left\langle 0\left|\hat{\sigma}_{i}\right| 0\right\rangle+\left|C_{1}(0)\right|^{2}\left\langle 1\left|\hat{\sigma}_{i}\right| 1\right\rangle+C_{0}^{*}(0) C_{1}(0) e^{-i \omega_{q} t}\left\langle 0\left|\hat{\sigma}_{i}\right| 1\right\rangle + C_{0}(0) C_{1}^{*}(0) e^{i \omega_{q} t}\left\langle 1\left|\hat{\sigma}_{i}\right| 0\right\rangle$


$\left\langle 0\left|\hat{\sigma}_{x}\right| 0\right\rangle=\left\langle 1\left|\hat{\sigma}_{x}\right| 1\right\rangle=\left\langle 0\left|\hat{\sigma}_{y}\right| 0\right\rangle=\left\langle 1\left|\hat{\sigma}_{y}\right| 1\right\rangle=0$

$\left\langle 0\left|\hat{\sigma}_{x}\right| 1\right\rangle=\left\langle 1\left|\hat{\sigma}_{x}\right| 0\right\rangle=1 \quad \& \quad\left\langle 0\left|\hat{\sigma}_{y}\right| 1\right\rangle=-\left\langle 1\left|\hat{\sigma}_{y}\right| 0\right\rangle=i$

$\left\langle\hat{\sigma}_{x}\right\rangle=C_{0}^{*}(0) C_{1}(0) e^{-i \omega_{4} t}+C_{1}^{*}(0) C_{0}(0) e^{i w_{q} t}=2 \operatorname{Re}\left(C_{0}^{*}(0) C_{1}(0) e^{-i \omega_{q} t}\right)$

$\left\langle\hat{\sigma}_{y}\right\rangle=i\left(C_{0}^{*}(0) C_{1}(0) e^{-i m_{4} t}+C_{1}^{*}(0) C_{0}(0) e^{i w_{0} t}\right)=2 R_{e}\left(i\left(_{0}^{*}(0) C_{1}(0) e^{-i w_{q} t}\right)\right.$

$|x+\rangle=\frac{1}{\sqrt{2}}(|1\rangle+|0\rangle) \quad \& \quad|x-\rangle=\frac{1}{\sqrt{2}}(|1\rangle-|0\rangle)$

$P_{x \pm}(t)=|\langle x \pm \mid \psi(t)\rangle|^{2}$

$\langle\pm \mid \psi(t)\rangle=\frac{1}{\sqrt{2}}\left(\pm C_{0}(0) e^{i \omega_{q} t / 2}\langle 0 \mid 0\rangle+C_{1}(0) e^{-i \omega_{9}^{\dagger} t / 2}\langle 1 \mid 1\rangle\right)$

$P_{x \pm}(t)=\frac{1}{2}\left(\left|C_{0}(0)\right|^{2}+\left|C_{1}(0)\right|^{2} \pm\left(C_{0}^{*} \cos \alpha_{1}(0) e^{-i w_{9} t}+C_{0} \cdot(0) C_{1}^{*}[0) e^{i \omega_{4} t}\right)\right)$

$=\frac{1}{2}\left(\left|C_{1}(0)\right|^{2}+\left|C_{0}(0)\right|^{2} \pm 2 \operatorname{Re}\left(C_{0}(0) C_{1}^{*}(0) e^{i \omega_{q} t}\right)\right)$

Unitary time evolution operator $|\psi(t)\rangle=U(t)|\psi(0)\rangle \rightarrow i \frac{d}{d t} U(t)=\hat{H} u(t)$

$U(t)=e^{-i \hat{H} t}$

$U(t)=\left[\begin{array}{ll}e^{-i \omega_{0} t / 2} & 0 \\ 0 & e^{i \omega_{q} t / 2}\end{array}\right]$

# Master equation

$\dot{\hat{\rho}}=\hat{\mathcal{L}} \hat{\rho}=-i\left[\hat{H}_{1} \hat{p}\right]+\sum_{i=1}^{n-1} k_{i}\left(\hat{F}_{i} \hat{p} F_{i}^{+}-\frac{1}{2}\left\{\hat{F}_{i}^{+} \hat{F}_{i}, \hat{\rho}\right\}\right)$

```python
frequency = 1
initialC0real = 0.5
initialC0imag = 0.5
initialC1real = 0.5
initialC1imag = 0.5

initialC0 = (initialC0real + 1j*initialC0imag)/((initialC0real**2) + (initialC0imag**2))
initialC1 = (initialC1real + 1j*initialC1imag)/((initialC1real**2) + (initialC1imag**2))

sz = qt.sigmaz()
sy = qt.sigmay()
sx = qt.sigmax()

ket1 = qt.basis(2, 0)
ket0 = qt.basis(2, 1)

qubitHamiltonian = 0.5*frequency*sz

initialState = qt.superPos(2, {0: initialC1, 1: initialC0}, populations=False)
initialC0 = initialState.A[1][0]
initialC1 = initialState.A[0]
```

```python
analyticalC0 = lambda t : initialC0*(np.e**(0.5*2*np.pi*1j*frequency*t))
analyticalC1 = lambda t : initialC1*(np.e**(-0.5*2*np.pi*1j*frequency*t))

expectationsAnalytic = {'sz':[], 'sx':[], 'sy':[]}
sxExpectation1 = lambda t: 2*(np.conj(initialC0)*initialC1*np.e**(-1j*2*np.pi*frequency*t)).real
szExpectation2 = (abs(initialC1)**2) - (abs(initialC0)**2)
```

```python
stepSize = 0.01
finalTime = 2 + stepSize
stepCount = int(finalTime / stepSize)

unitaryEvolution = qt.Unitary(2*np.pi*qubitHamiltonian, stepSize)
stepDiff = -1j*2*np.pi*qubitHamiltonian*stepSize
```

```python
numericalTimeC_0real = []
numericalTimeC_0imag = []
numericalTimeC_1real = []
numericalTimeC_1imag = []

szExpectation = []
sxExpectation = []
syExpectation = []

state = initialState
for i in range(stepCount):
    ct1 = qt.innerProd(state, ket1) # <ket1|state> ordering is important
    ct0 = qt.innerProd(state, ket0)

    numericalTimeC_0real.append(ct0.real)
    numericalTimeC_0imag.append(ct0.imag)

    numericalTimeC_1real.append(ct1.real)
    numericalTimeC_1imag.append(ct1.imag)

    szExpectation.append(qt.expectation(sz, state))
    syExpectation.append(qt.expectation(sy, state))
    sxExpectation.append(qt.expectation(sx, state))

    expectationsAnalytic['sx'].append(sxExpectation1(i*stepSize))
    expectationsAnalytic['sz'].append(szExpectation2)

    state = unitaryEvolution @ state
    #state += stepDiff @ state
```

```python
timeList = [i*stepSize for i in range(stepCount)]

C_0Analytical = [analyticalC0(i*stepSize) for i in range(stepCount)]
C_0Analyticalr = [c.real for c in C_0Analytical]
C_0Analyticali = [c.imag for c in C_0Analytical]

C_1Analytical = [analyticalC1(i*stepSize) for i in range(stepCount)]
C_1Analyticalr = [c.real for c in C_1Analytical]
C_1Analyticali = [c.imag for c in C_1Analytical]
```
